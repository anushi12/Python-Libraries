```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(1, 10, 200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.figure(figsize=(6,2))
plt.plot(x,y, '--g')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.grid()
plt.show()
```

```python
a  = np.random.rand(50)
b  = np.random.rand(50)
plt.scatter(a,b)
```

```python
## to change color (c), sizes (s)and aplha
colors = np.random.rand(50)
sizes = 100* np.random.rand(50)   # multiply with any number is the sizes are too big
plt.scatter(a, b, c = colors, s = sizes, alpha=.5)
```

```python
c = ['a', 'b', 'c', 'd', 'e']
d = np.random.rand(5)
plt.bar(c,d)
```

```python
data = [1,3,3,4,5,3,7,3,6,2,1,5,6,7,9,4,3,2,3,6,8,9,9,4,3,2,4,7,8]
plt.hist(data)
plt.xlabel("data")
plt.ylabel("frequency")
plt.title("Histogram")
plt.show()
```

```python
g = np.random.rand(50)
h = np.random.rand(50)
i = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(g,h,i)
ax.set_xlabel("g axis")
ax.set_ylabel("h axis")
ax.set_zlabel("i axis")
```

```python
fruits = np.array(['45', '12', '7', '36'])
my_labels = ['apples', 'blueberry', 'strawberry', 'watermelon']
colors = ['violet', 'yellow', 'pink', 'lightgreen', 'brown']
plt.pie(fruits, labels = my_labels, startangle= 90, colors = colors, shadow= True)
plt.legend(title = "Fruits")
plt.show()
```
