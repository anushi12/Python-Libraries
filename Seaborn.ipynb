```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.scatterplot(x = iris.sepal_length, y = iris.petal_width)
```

```python
sns.displot(iris['sepal_length'])   # displot is equivalent to histogram
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x =tips.total_bill, y = tips.tip)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, hue = 'smoker', data = tips)     ## relplot is used to draw the complex relationship easier
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, hue = 'size', data = tips)   ## style can be used instead of hue which groups data based on different shapes eg: style = 'size'
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, hue='time', style='size', data = tips)
```

```python
sns.catplot(y = tips.total_bill, x = tips.day, data = tips) # categorical plot
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip, data = tips)
```

```python
sns.pairplot(tips)
```
